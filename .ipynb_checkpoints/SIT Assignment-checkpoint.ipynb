# After applying principal component analysis (PCA) on a given dataset, it was found that the percentage of variance for the first N components is X%. How is this percentage of variance computed?

```python
kmeans = KMeans(n_clusters=7) #n_clusters define the number of clusters to find
kmeans.fit(q1) # we use .fit() to fit the data (learning)
```

```python
centroids = kmeans.cluster_centers_
labels    = kmeans.labels_

colors=["g.","r."]

```

```python
y = kmeans.fit_predict(q1)

q1['Cluster_predicted'] = y

q1
```

```python
# Calculating Eigenvectors and eigenvalues of Cov matirx
mean_vector = np.mean(q1, axis=0)
cov_matrix = np.cov(q1.T)
eig_values, eig_vectors = np.linalg.eig(cov_matrix)
```

```python
# Create a list of (eigenvalue, eigenvector) tuples
eigen_pairs = [ (np.abs(eig_values[i]),eig_vectors[:,i]) for i in range(len(eig_values))]

# Sort from high to low
eigen_pairs.sort(key = lambda x: x[0], reverse= True)

# Calculation of Explained Variance from the eigenvalues
total = sum(eig_values)
var_explained = [(i/total)*100 for i in sorted(eig_values, reverse=True)] # Individual explained variance
cum_var_explained = np.cumsum(var_explained) # Cumulative explained variance
```

```python
print(cum_var_explained)
print(len(cum_var_explained))
```

```python
plt.figure(figsize=(10, 5))
plt.bar(range(141), var_explained, alpha=0.3333, align='center', label='individual explained variance', color = 'g')
plt.step(range(141), cum_var_explained, where='mid',label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.show()
```
